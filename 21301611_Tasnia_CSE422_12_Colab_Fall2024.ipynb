**Feature selection**

```python
car = car[car['Mileage Unit'] != 'km/kg']
car
```

**Feature Encoding**

```python
car.info()
```

```python
## feature encoding
# Transform the category_desc column
name_enc = pd.get_dummies(car['name'],dtype='int')
owner_enc = pd.get_dummies(car['owner'],dtype='int')
car = pd.concat([car, name_enc], axis=1)
car = pd.concat([car, owner_enc], axis=1)

from sklearn.preprocessing import LabelEncoder
enc_transmission = LabelEncoder()
car['transmission_enc'] = enc_transmission.fit_transform(car['transmission'])

enc_seller_type = LabelEncoder()
car['seller_type_enc'] = enc_seller_type.fit_transform(car['seller_type'])

enc_fuel = LabelEncoder()
car['fuel_enc'] = enc_fuel.fit_transform(car['fuel'])
```

```python
car.head()
```

```python
car.info()
```

```python
car = car.drop(['name', 'owner', 'transmission', 'seller_type', 'fuel'], axis = 1)
car.info()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
car['selling_price'].hist(bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Selling Price')
plt.ylabel('Number of Instances')
plt.title('Distribution of Selling Prices')
plt.show()
```

```python

threshold = 1100000
car[car['selling_price'] > threshold]
```

```python
car = car[car['selling_price'] <= threshold]
```

```python
from sklearn.model_selection import train_test_split
X = car.drop('selling_price', axis=1)  # Features
y = car['selling_price']              # Labels

# Step 3: Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Step 4: Print the shapes of the splits
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
y_train=y_train/10000
y_test=y_test/10000
```

**Model implementation**

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train, y_train)
yprediction_lr = clf.predict(X_test)
```

```python
from sklearn import tree
clf1 = tree.DecisionTreeRegressor()
clf1 = clf1.fit(X_train, y_train)
yprediction_dt=clf1.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor = rf_regressor.fit(X_train, y_train)
yprediction_rf = rf_regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np


# Mean Squared Error (MSE)
mse_lr = mean_squared_error(y_test, yprediction_lr)
mse_dt = mean_squared_error(y_test, yprediction_dt)
mse_rf = mean_squared_error(y_test, yprediction_rf)


# Root Mean Squared Error (RMSE)
rmse_lr = np.sqrt(mse_lr)
rmse_dt = np.sqrt(mse_dt)
rmse_rf = np.sqrt(mse_rf)



# Print the results
print(f"Mean Squared Error LR(MSE): {mse_lr:.2f}")
print(f"Mean Squared Error DT(MSE): {mse_dt:.2f}")
print(f"Mean Squared Error RF(MSE): {mse_rf:.2f}")

print(f"Root Mean Squared Error LR(RMSE): {rmse_lr:.2f}")
print(f"Root Mean Squared Error DT(RMSE): {rmse_dt:.2f}")
print(f"Root Mean Squared Error RF(RMSE): {rmse_rf:.2f}")
```

```python
from sklearn import metrics
# R² Score

r2_lr = metrics.r2_score(y_test, yprediction_lr)
r2_dt = metrics.r2_score(y_test, yprediction_dt)
r2_rf = metrics.r2_score(y_test, yprediction_rf)

print(f"R Squared Error LR(R²): {r2_lr:.2f}")
print(f"R Squared Error DT(R²): {r2_dt:.2f}")
print(f"R Squared Error RF(R²): {r2_rf:.2f}")
```

```python
from sklearn.metrics import mean_absolute_error

mae_lr = mean_absolute_error(y_test, yprediction_lr)
mae_dt = mean_absolute_error(y_test, yprediction_dt)
mae_rf = mean_absolute_error(y_test, yprediction_rf)

print("Mean Absolute Error LR(MAE):", mae_lr)
print("Mean Absolute Error DT(MAE):", mae_dt)
print("Mean Absolute Error RF(MAE):", mae_rf)
```

```python

import matplotlib.pyplot as plt

# RMSE values for the three models
rmse_values = [rmse_lr, rmse_dt, rmse_rf]
models = ['Linear Regression', 'Decision Tree', 'Random Forest']

plt.figure(figsize=(8, 6))
plt.bar(models, rmse_values, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel("Models")
plt.ylabel("Root Mean Squared Error (RMSE)")
plt.title("RMSE Comparison of Top Three Models")

for i, v in enumerate(rmse_values):
    plt.text(i, v + 0.1, str(round(v, 2)), ha='center')
plt.show()
```
